**5회 반복 출력**

**i부터 n까지의 합(3단 전략)**

```python
sum = 0
i = 1
n = int(input("숫자를 입력해주세요 : "))

while i <= n :
    sum += i
    i += 1
# for i in range(1, n + 1) :
#     sum += i

print("1부터 " + str(n) + "까지의 합 : " + str(sum))
```

**사용자가 입력한 단의 구구단 출력 문제**

```python
base = int(input("출력할 단을 입력하세요: "))
i = 1

while  i <= 9 :
    print(str(base) + " X " + str(i) + " = " + str(base * i))
    i += 1
```

# 6강 2절

**반지름은 10이고 높이가 1, 5, 14, 26, 31인 원뿔의 부피와 겉넓이를 각각 출력하는 문제**

```python
# 리스트 생성
hei_list = [1, 4, 14, 26, 31]

for hei in hei_list :
    sla = pow(10**2 + hei**2, 0.5)
    vol = 1/3 * 3.14 * 10**2 * hei
    suf = 3.14 * 10**2 + 3.14 * 10 * sla
    print("부피: " + str(vol) + " 겉넓이: " + str(suf))
```

**유클리드 호제법**

```python
# 유클리드 호제법으로 두 수의 최대공약수(GCD)를 구하는 프로그램

# 두 수 입력 받기
a = int(input("첫 번째 숫자를 입력하세요: "))
b = int(input("두 번째 숫자를 입력하세요: "))

# a를 b로 나눈 나머지를 b에 저장
# b가 0이 될 때까지 반복

while b != 0 :
    rem = a % b
    a = b
    b = rem

print("두 수의 최대공약수는:", a)
```
